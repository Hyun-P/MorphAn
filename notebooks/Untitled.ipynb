```python
import os
import shutil
```

```python
parent = "/data/hyun/Analysis/data/UB.VI.220"
```

```python
for root,dirs,files in os.walk(parent):
    if not files: continue
    if os.path.basename(root) == 'orig': continue
    gparent = os.path.dirname(root)
    print(gparent)
    print(root)
    for f in files:
        if '.tif' in f:
            nn = f.split('_MMStack')[0] + '.tif'
            opath = os.path.join(root,f)
            print(nn)
    
    div = os.path.basename(gparent).split('_')[0]
    ngparent = os.path.join(parent,div)
    if not os.path.exists(ngparent):
        os.mkdir(ngparent)
        
    ngparent2 = os.path.join(ngparent,'orig')
    if not os.path.exists(ngparent2):
        os.mkdir(ngparent2)
        
    npath = os.path.join(ngparent2, nn)
    shutil.copyfile(opath,npath)
    
    print(ngparent)
    print(ngparent2)
    print(npath)
    print()
```
