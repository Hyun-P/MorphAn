```python
MA = MorphAn()

# a path to the experiment
# pipeline_dir = os.path.join(module_path,'src','example','analysis_pipeline')
pipeline_dir = '/data/hyun/Analysis/data/UB.VI.220/DIV5'
    
swc_data_dir = os.path.join(pipeline_dir,'swc_paths_data')

recon_dir = MA.check_directory(pipeline_dir,'reconstruction')

sholl_path = os.path.join(pipeline_dir,'sholl.csv')
```

```python
path_dict = {}
for var in os.listdir(swc_data_dir):
    print('Preprocessing... %s' %var)
    neuron_name = var.split('.csv')[0]
    
    data_path = os.path.join(swc_data_dir,var)

    tmp_dict = MA.iron_preprocess(data_path)
    path_dict[neuron_name] = tmp_dict

path_dict = MA.adoption_agency(path_dict)
print("Preprocess complete.\n\n")

tmp_list = []
for k,v in path_dict.items():
    print("Reconstructing... %s"%k)
    
    reconstructed_neuron = MA.reconstruct_neuron(v)
    output = os.path.join(recon_dir,'%s.tif'%k)
    io.imsave(output, reconstructed_neuron)
    
    dft = MA.iron_sholl(reconstructed_neuron)
    dft['Neuron_Name'] = k
    tmp_list.append(dft)

print("Reconstruction complete.\n\n")
    
df = pd.concat(tmp_list)
    
    
df.to_csv(sholl_path)
df
```
